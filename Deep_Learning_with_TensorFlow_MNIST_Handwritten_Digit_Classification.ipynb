```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def load_and_preprocess_data():
    """Load and preprocess the MNIST dataset"""
    print("=" * 60)
    print("LOADING AND PREPROCESSING MNIST DATASET")
    print("=" * 60)

    # Load MNIST dataset
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

    print("Original data shapes:")
    print(f"Training images: {x_train.shape}")
    print(f"Training labels: {y_train.shape}")
    print(f"Test images: {x_test.shape}")
    print(f"Test labels: {y_test.shape}")

    print(f"\nPixel value range: {x_train.min()} to {x_train.max()}")
    print(f"Number of classes: {len(np.unique(y_train))}")

    # Normalize pixel values to [0, 1] range
    print("\nNormalizing pixel values...")
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

    # Reshape data to add channel dimension for CNN
    print("Reshaping data for CNN input...")
    x_train = x_train.reshape(-1, 28, 28, 1)
    x_test = x_test.reshape(-1, 28, 28, 1)

    # Convert labels to categorical (one-hot encoding)
    print("Converting labels to categorical format...")
    y_train_cat = keras.utils.to_categorical(y_train, 10)
    y_test_cat = keras.utils.to_categorical(y_test, 10)

    print("\nPreprocessed data shapes:")
    print(f"Training images: {x_train.shape}")
    print(f"Training labels: {y_train_cat.shape}")
    print(f"Test images: {x_test.shape}")
    print(f"Test labels: {y_test_cat.shape}")

    # Display sample images
    visualize_sample_data(x_train, y_train)

    return (x_train, y_train, y_train_cat), (x_test, y_test, y_test_cat)

def visualize_sample_data(x_train, y_train, num_samples=10):
    """Visualize sample images from the dataset"""
    print(f"\nDisplaying {num_samples} sample images...")

    plt.figure(figsize=(15, 3))
    for i in range(num_samples):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(x_train[i].reshape(28, 28), cmap='gray')
        plt.title(f'Label: {y_train[i]}')
        plt.axis('off')

    plt.suptitle('Sample MNIST Images', fontsize=16)
    plt.tight_layout()
    plt.show()

def build_cnn_model():
    """Build and compile the CNN model"""
    print("\n" + "=" * 60)
    print("BUILDING CNN MODEL ARCHITECTURE")
    print("=" * 60)

    model = keras.Sequential([
        # First Convolutional Block
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1),
                     name='conv2d_1'),
        layers.BatchNormalization(name='batch_norm_1'),
        layers.MaxPooling2D((2, 2), name='maxpool_1'),
        layers.Dropout(0.25, name='dropout_1'),

        # Second Convolutional Block
        layers.Conv2D(64, (3, 3), activation='relu', name='conv2d_2'),
        layers.BatchNormalization(name='batch_norm_2'),
        layers.MaxPooling2D((2, 2), name='maxpool_2'),
        layers.Dropout(0.25, name='dropout_2'),

        # Third Convolutional Block
        layers.Conv2D(128, (3, 3), activation='relu', name='conv2d_3'),
        layers.BatchNormalization(name='batch_norm_3'),
        layers.Dropout(0.25, name='dropout_3'),

        # Dense Layers
        layers.Flatten(name='flatten'),
        layers.Dense(512, activation='relu', name='dense_1'),
        layers.BatchNormalization(name='batch_norm_4'),
        layers.Dropout(0.5, name='dropout_4'),
        layers.Dense(10, activation='softmax', name='output')
    ])

    # Compile the model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("Model Architecture:")
    model.summary()

    # Count total parameters
    total_params = model.count_params()
    print(f"\nTotal trainable parameters: {total_params:,}")

    return model

def train_model(model, train_data, test_data, epochs=12):
    """Train the CNN model"""
    print("\n" + "=" * 60)
    print("TRAINING THE MODEL")
    print("=" * 60)

    x_train, y_train, y_train_cat = train_data
    x_test, y_test, y_test_cat = test_data

    # Setup callbacks
    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=3,
            restore_best_weights=True,
            verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=2,
            min_lr=0.0001,
            verbose=1
        )
    ]

    print(f"Starting training for up to {epochs} epochs...")
    print("Callbacks: Early Stopping, Learning Rate Reduction")

    # Train the model
    history = model.fit(
        x_train, y_train_cat,
        batch_size=128,
        epochs=epochs,
        validation_data=(x_test, y_test_cat),
        callbacks=callbacks,
        verbose=1
    )

    print("\nTraining completed!")
    return history

def evaluate_model(model, test_data):
    """Evaluate the trained model"""
    print("\n" + "=" * 60)
    print("MODEL EVALUATION")
    print("=" * 60)

    x_test, y_test, y_test_cat = test_data

    # Evaluate on test set
    test_loss, test_accuracy = model.evaluate(x_test, y_test_cat, verbose=0)

    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

    # Make predictions
    print("\nGenerating predictions...")
    y_pred_proba = model.predict(x_test, verbose=0)
    y_pred = np.argmax(y_pred_proba, axis=1)

    # Detailed classification report
    print("\nDetailed Classification Report:")
    print(classification_report(y_test, y_pred,
                              target_names=[str(i) for i in range(10)]))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(cm)

    return test_accuracy, y_pred, y_pred_proba

def visualize_predictions(model, test_data, num_samples=5):
    """Visualize model predictions on sample images"""
    print(f"\n" + "=" * 60)
    print(f"VISUALIZING PREDICTIONS ON {num_samples} SAMPLE IMAGES")
    print("=" * 60)

    x_test, y_test, y_test_cat = test_data

    # Get random samples
    np.random.seed(42)  # For reproducible results
    indices = np.random.choice(len(x_test), num_samples, replace=False)

    # Make predictions on selected samples
    sample_images = x_test[indices]
    sample_labels = y_test[indices]
    predictions = model.predict(sample_images, verbose=0)
    predicted_classes = np.argmax(predictions, axis=1)
    confidence_scores = np.max(predictions, axis=1)

    # Create visualization
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))

    for i in range(num_samples):
        # Display image
        axes[0, i].imshow(sample_images[i].reshape(28, 28), cmap='gray')
        axes[0, i].set_title(f'True: {sample_labels[i]}\nPred: {predicted_classes[i]}\nConf: {confidence_scores[i]:.3f}')
        axes[0, i].axis('off')

        # Display prediction probabilities
        axes[1, i].bar(range(10), predictions[i], alpha=0.7,
                      color='skyblue', edgecolor='navy')
        axes[1, i].set_title(f'Prediction Probabilities')
        axes[1, i].set_xlabel('Digit Class')
        axes[1, i].set_ylabel('Probability')
        axes[1, i].set_xticks(range(10))

        # Highlight predicted class
        axes[1, i].bar(predicted_classes[i], predictions[i][predicted_classes[i]],
                      color='red', alpha=0.8)

    plt.suptitle('CNN Model Predictions on MNIST Test Images', fontsize=16)
    plt.tight_layout()
    plt.show()

    # Print detailed results
    print("Prediction Details:")
    for i, idx in enumerate(indices):
        print(f"Sample {i+1}: True={sample_labels[i]}, "
              f"Predicted={predicted_classes[i]}, "
              f"Confidence={confidence_scores[i]:.3f}")

def plot_training_history(history):
    """Plot training history"""
    print("\n" + "=" * 60)
    print("TRAINING HISTORY VISUALIZATION")
    print("=" * 60)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot accuracy
    ax1.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    ax1.set_title('Model Accuracy Over Time')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim([0.9, 1.0])

    # Plot loss
    ax2.plot(history.history['loss'], label='Training Loss', linewidth=2)
    ax2.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    ax2.set_title('Model Loss Over Time')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Print final metrics
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    print(f"Final Training Accuracy: {final_train_acc:.4f}")
    print(f"Final Validation Accuracy: {final_val_acc:.4f}")

def create_confusion_matrix_heatmap(y_true, y_pred):
    """Create a detailed confusion matrix heatmap"""
    print("\nCreating confusion matrix heatmap...")

    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(10), yticklabels=range(10),
                cbar_kws={'label': 'Count'})
    plt.title('Confusion Matrix - MNIST CNN Classification')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

    # Calculate per-class accuracy
    class_accuracy = cm.diagonal() / cm.sum(axis=1)
    print("\nPer-class Accuracy:")
    for i, acc in enumerate(class_accuracy):
        print(f"Digit {i}: {acc:.4f} ({acc*100:.2f}%)")

def main():
    """Main function to run the complete MNIST CNN pipeline"""
    print("MNIST HANDWRITTEN DIGIT CLASSIFICATION WITH CNN")
    print("=" * 80)

    try:
        # Set random seeds for reproducibility
        tf.random.set_seed(42)
        np.random.seed(42)

        # Step 1: Load and preprocess data
        train_data, test_data = load_and_preprocess_data()

        # Step 2: Build model
        model = build_cnn_model()

        # Step 3: Train model
        history = train_model(model, train_data, test_data)

        # Step 4: Evaluate model
        test_accuracy, y_pred, y_pred_proba = evaluate_model(model, test_data)

        # Step 5: Visualize predictions
        visualize_predictions(model, test_data)

        # Step 6: Plot training history
        plot_training_history(history)

        # Step 7: Create confusion matrix
        _, y_test, _ = test_data
        create_confusion_matrix_heatmap(y_test, y_pred)

        # Final summary
        print("\n" + "=" * 80)
        print("FINAL SUMMARY")
        print("=" * 80)
        print(f"✓ Model trained successfully!")
        print(f"✓ Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

        if test_accuracy > 0.95:
            print("✓ SUCCESS: Achieved >95% test accuracy requirement!")
        else:
            print("⚠ WARNING: Did not achieve >95% test accuracy. Consider:")
            print("  - Training for more epochs")
            print("  - Adjusting learning rate")
            print("  - Adding more data augmentation")

        print("✓ Prediction visualizations created successfully!")
        print("✓ Analysis completed!")

        # Save model
        model.save('mnist_cnn_model.h5')
        print("✓ Model saved as 'mnist_cnn_model.h5'")

    except Exception as e:
        print(f"Error occurred: {str(e)}")
        print("Please check your TensorFlow installation and dependencies.")

if __name__ == "__main__":
    main()
```
